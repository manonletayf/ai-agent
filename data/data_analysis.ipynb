```python
import pandas as pd
import csv
import os
```

```python
csv.field_size_limit(10**7)

input_path = 'corporate_uk/companies.csv'
temp_path = 'corporate_uk/companies_temp.csv'
final_path = 'corporate_uk/companies_final.csv'

# Step 1: Clean up double commas and ', ,' in batches
batch_size = 10000
with open(input_path, 'r', encoding='utf-8') as infile, \
     open(temp_path, 'w', encoding='utf-8', newline='') as outfile:
    batch = []
    for i, line in enumerate(infile, 1):
        # Replace all ',,' and ', ,' with ','
        while ',,' in line or ', ,' in line:
            line = line.replace(',,', ',').replace(', ,', ',')
        batch.append(line)
        if i % batch_size == 0:
            outfile.writelines(batch)
            batch = []
    if batch:
        outfile.writelines(batch)

# Step 2: Normalize columns
with open(temp_path, 'r', encoding='utf-8') as infile:
    reader = csv.reader(infile)
    header = next(reader)
    num_columns = len(header)

with open(temp_path, 'r', encoding='utf-8') as infile, \
     open(final_path, 'w', encoding='utf-8', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    writer.writerow(next(reader))  # write header

    for row in reader:
        if len(row) < num_columns:
            row += [''] * (num_columns - len(row))
        elif len(row) > num_columns:
            row = row[:num_columns-1] + [','.join(row[num_columns-1:])]
        writer.writerow(row)

# Step 3: Load with pandas
df = pd.read_csv(final_path)
```
